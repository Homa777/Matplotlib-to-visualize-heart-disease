## Exploring data with Pandas - visualize data of heartdisease dataset

```python
import pandas as pd

# Load data from a text file

df_heart = pd.read_csv('heart_2022_Key_indicators.csv',delimiter=',',header='infer')

df_heart
```

```python
# Remove any rows with missing data
df_heart = df_heart.dropna(axis=0, how='any')

```

## Add a column of patients with heartdisease regardless of gender

```python
passes_total  = pd.Series(df_heart['HeartDisease']=="Yes")
df_heart= pd.concat([df_heart, passes_total.rename("total_patient_heartdisease")], axis=1)
df_heart
```

## Add a column of females who have heartdisease

```python
passes  = pd.Series((df_heart['Sex'] == "Female") & (df_heart['HeartDisease']=="Yes"))
df_heart= pd.concat([df_heart, passes.rename("female_with_heartdisease")], axis=1)
df_heart
```

```python
# Create a figure for 2 subplots (1 row, 2 columns)

from matplotlib import pyplot as plt

fig, ax = plt.subplots(1, 2, figsize = (10,4))

# Create a bar plot of name vs grade on the first axis
pass_counts = df_heart['HeartDisease'].value_counts()
ax[0].pie(pass_counts, labels=pass_counts)
ax[0].set_title('number of patients with HeartDisease ')
ax[0].legend(pass_counts.keys().tolist())

# Create a pie chart of pass counts on the second axis
pass_counts = df_heart['female_with_heartdisease'].value_counts()
ax[1].pie(pass_counts, labels=pass_counts)
ax[1].set_title('number of female patient with HeartDisease ')
ax[1].legend(pass_counts.keys().tolist())

# Show the figure
fig.show()
```

```python
# This shows the number of female with heart disease almost same; they are slighlty less 
```
